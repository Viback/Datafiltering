```python
import bs4 #beautifulsoup 4
import csv
from urllib.request import urlopen as uReq #url open module
from bs4 import BeautifulSoup as soup #beautifulsoup module
import numpy as np # imports a fast numerical programming library
import scipy as sp #imports stats functions, amongst other things
import matplotlib as mpl # this actually imports matplotlib
import matplotlib.cm as cm #allows us easy access to colormaps
import matplotlib.pyplot as plt #sets up plotting under plt
import pandas as pd #lets us handle data as dataframes
#sets up pandas table display
pd.set_option('display.width', 500)
pd.set_option('display.max_columns', 100)
pd.set_option('display.notebook_repr_html', True)
import seaborn as sns #sets up styles and gives us more plotting options
```

```python
#Read in ratings data
col1 = ['user_id', 'movie_id', 'rating', 'unix_timestamp']
ratings = pd.read_csv('u.data', sep = '\t', names = col1)
ratings.head()
```

```python
#Inläsning av film data för del 1 samt del 2 avgränsning av kolumner
col2 = ['movie_id', 'title', 'release_date', 'video_release_date', 'imdb_url']
movies = pd.read_csv('u.item', sep = '|', names = col2, usecols = range(5), encoding = "iso-8859-1")
movies.head()
```

```python
#Inläsning av data för del 1 samt del 2 avgränsning av kolumner
col3 = ['user_id', 'age', 'gender', 'occupation', 'zip']
users = pd.read_csv('u.user', sep = '|', names = col3, usecols = range(4))
users.head()
#All data i en dataframe
movie_ratings = pd.merge(movies, ratings)
lens = pd.merge(movie_ratings, users)
```

```python
#Filtrering av data enlight yrke, ålder eller kön
users.query("age > 40")
```

```python
#Filtrering av data enlight yrke, ålder och kön för del 3
users[(users.age > 40) & (users.gender == "M") & (users.occupation == "scientist")]
```

```python
#Del 4, utforska data. Medelåldern av författare.
writers = users[(users.occupation == "writer")]
np.mean(writers.age)
```

```python
#Medelrating per film
average = movies[['movie_id', 'title']].set_index('movie_id').join(
    ratings[['movie_id','rating']].groupby('movie_id').mean())
average
```

```python
#Ratings per film och medeltal
movie_stats = lens.groupby('title').agg({'rating': [np.size, np.mean]})
movie_stats.head()
```

```python
#Top 10 filmerna med över 40 ratings
movie_stats.sort_values([('rating', 'mean')], ascending=False).head()
atleast_40 = movie_stats['rating']['size'] >= 40
movie_stats[atleast_40].sort_values([('rating', 'mean')], ascending=False)[:10]
```

```python
#medelrating per användare
useraverage = lens.groupby('user_id').agg({'rating': [np.size, np.mean]})
useraverage
```

```python
#Tuffaste och snällaste
print(useraverage.rating.sort_values(by=['mean'])[:5])
print(useraverage.rating.sort_values(by=['mean'])[-5:])
```
